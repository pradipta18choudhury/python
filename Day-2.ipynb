##  Handling NAN

```python
# we have cinverted two rows as NAN
df.iloc[2:4,1:3]=np.nan
df.describe()
```

```python
df.describe()
```

```python
# if we want to drop the rows containing nan
df.dropna(inplace=True)
# we see 2 and 3 row not present in dataset
df.head()
```

```python
df.head()
# we have reset the dropped values again
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
# let suppose we want to fill the NAN places
df.iloc[2:4,1:3]=np.nan
df.head()
```

```python
# suppose we want to replace or fill the nan values with mean values of the whole column
df.sw.fillna(df.sw.mean(),inplace=True)
df.pl.fillna(df.pl.mean(),inplace=True)
df.head()
```
